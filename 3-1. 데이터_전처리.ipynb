```python
from sklearn import preprocessing  
```

```python
nba_player_of_the_week_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Study/Code_It/ML/3. 머신러닝, 더 빠르고 정확하게/NBA_player_of_the_week.csv")
nba_player_of_the_week_df
```

```python
height_weight_age_df = nba_player_of_the_week_df[['Height CM','Weight KG', 'Age']]
height_weight_age_df.head()
```

```python
scaler = preprocessing.MinMaxScaler()
normalized_data = scaler.fit_transform(height_weight_age_df)
```

```python
normalized_data
```

```python
normalized_df = pd.DataFrame(normalized_data, columns=['Height', 'Weight', 'Age'])
```
